```python
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Prepare the data
dates = pd.to_datetime(filtered_data['Amount'])
values =filtered_data['Amount']
series = pd.Series(values.values, index=dates)

# Determine the optimal values for p and q using ACF and PACF plots
plot_acf(series)
plt.show()

plot_pacf(series)
plt.show()

# Grid search to find the optimal values for p, d, and q
best_aic = np.inf
best_order = None

for p in range(3):
    for d in range(2):
        for q in range(3):
            try:
                model = ARIMA(series, order=(p, d, q))
                model_fit = model.fit()
                aic = model_fit.aic

                if aic < best_aic:
                    best_aic = aic
                    best_order = (p, d, q)
            except:
                continue
print(f"Optimal (p, d, q) values: {best_order}")
```

```python
from statsmodels.tsa.arima.model import ARIMA

# Prepare the data
dates = pd.to_datetime(filtered_data['Amount'])
values = filtered_data['Amount']
series = pd.Series(values.values, index=dates)

# Fit the ARIMA model
order = (1,1,1)
model = ARIMA(filtered_data['Amount'], order=order)
model_fit = model.fit()

print(model_fit.summary())
predictions = model_fit.predict()

# Forecast the next 39 values
forecast = model_fit.forecast(steps=39)

# Print the forecasted values
print(forecast)
```

```python
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error,mean_squared_error, mean_absolute_error
```

```python
mae = mean_absolute_error(filtered_data['Amount'], predictions)
print("MAE:", mae)
mape = mean_absolute_percentage_error(filtered_data['Amount'],predictions) * 100
print("MAPE:", mape)
```

```python
#Plot the results
plt.figure(figsize=(20,6))
plt.plot(filtered_data['Amount'], label='Actual',linewidth=2)
plt.plot(predictions, label='Predicted',linewidth=5)

plt.xlabel('Trend')
plt.ylabel('Value')
plt.title('ARIMA Model')
plt.legend()
plt.show()
```
